```python
# Se hace la lectura del dataset del seminario y se obtiene su cabecera
dt = pd.read_csv('C:/Users/MauricioR/Documents/Pagos/MatriculaVIU/Maestria_Big_Data_Data_Science_Sep_2019/11MBIG_TFM/Datos/Data_frame.csv', ',', index_col='ID')
dt.head()
```

```python
def word_count(str):
    counts = 0
    words = str.split()

    for word in words:        
        counts += 1
    return counts

#print( word_count('the quick brown fox jumps over the lazy dog.'))
dt['Word_count_work_description'] = 0
for ind in dt.index:
    dt['Word_count_work_description'][ind] = word_count(str(dt['Work_description'][ind]))
    #print( word_count(str(dt['Work_description'][ind])))
dt.head()
```
